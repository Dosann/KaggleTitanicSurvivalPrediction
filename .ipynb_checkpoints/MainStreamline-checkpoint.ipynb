```python
import pandas as pd, numpy as np
import tensorflow as tf
import copy, time, string
from sklearn import svm
from sklearn.decomposition import PCA
import xgboost as xgb
```

```python
class dataPool():
    def __init__(self, data, axis = 1):
        self.dataAxis = axis
        if self.dataAxis == 1:
            self.dataset = data
        else:
            self.dataset = data.T
        self.currentPoint = 0
        self.lastpoint = 9
        self.datasetSize = self.dataset.shape[1]
    
    def getData(self, size = None):
        if size == None:
            size = int(self.datasetSize / 10)
        self.lastPoint = self.currentPoint
        self.currentPoint += size
        block1 = self.dataset[:, self.lastPoint:self.currentPoint]
        block2 = self.dataset[:, 0:max([self.currentPoint-self.datasetSize, 0])]
        datablock = np.hstack([block1, block2])
        self.currentPoint = self.currentPoint % self.datasetSize
        return datablock
def calculatePrecision(inputX, inputY, correctFunction):
    return sess.run(correctFunction, feed_dict={X:inputX, Y_: inputY})
```

```python

dfTrainOriCount = int(dfTrainOri.shape[0]*0.8)
dfTrainOri = dfTrainOri.iloc[:dfTrainOriCount, :]
dfDevOri = dfTrainOri.iloc[dfTrainOriCount:, :]



```

```python
dfTrainY = dfTrainOri['Survived']

dfTrainX = copy.copy(dfTrainOri)
dfTrainX['SEmbarked'] = dfTrainX['Embarked'] == 'S'
dfTrainX['CEmbarked'] = dfTrainX['Embarked'] == 'C'
dfTrainX['QEmbarked'] = dfTrainX['Embarked'] == 'Q'
dfTrainX['nameLength'] = dfTrainX['Name'].apply(len)
dfTrainX.loc[dfTrainX['Sex'] == 'male', 'Sex'] = 1
dfTrainX.loc[dfTrainX['Sex'] == 'female', 'Sex'] = 0

dfTrainX = dfTrainX.drop(['PassengerId','Survived','Name','Ticket','Cabin','Embarked'], axis = 1)
dfTrainX.loc[np.isnan(dfTrainX['Age']), 'Age'] = np.mean(dfTrainX['Age'])

arTrainX = np.array(dfTrainX).T
arTrainY = np.array(dfTrainY).T.reshape([1, -1])
```

```python
dfDevY = dfDevOri['Survived']

dfDevX = copy.copy(dfDevOri)
dfDevX['SEmbarked'] = dfDevX['Embarked'] == 'S'
dfDevX['CEmbarked'] = dfDevX['Embarked'] == 'C'
dfDevX['QEmbarked'] = dfDevX['Embarked'] == 'Q'
dfDevX['nameLength'] = dfDevX['Name'].apply(len)
dfDevX.loc[dfTrainX['Sex'] == 'male', 'Sex'] = 1
dfDevX.loc[dfTrainX['Sex'] == 'female', 'Sex'] = 0

dfTrainX = dfTrainX.drop(['PassengerId','Survived','Name','Ticket','Cabin','Embarked'], axis = 1)
dfTrainX.loc[np.isnan(dfTrainX['Age']), 'Age'] = np.mean(dfTrainX['Age'])

arTrainX = np.array(dfTrainX).T
arTrainY = np.array(dfTrainY).T.reshape([1, -1])
```

```python
dfTestX = copy.copy(dfTestOri)
dfTestX['SEmbarked'] = dfTestX['Embarked'] == 'S'
dfTestX['CEmbarked'] = dfTestX['Embarked'] == 'C'
dfTestX['QEmbarked'] = dfTestX['Embarked'] == 'Q'
dfTestX['nameLength'] = dfTestX['Name'].apply(len)
dfTestX.loc[dfTestX['Sex'] == 'male', 'Sex'] = 1
dfTestX.loc[dfTestX['Sex'] == 'female', 'Sex'] = 0

dfTestX = dfTestX.drop(['PassengerId','Name','Ticket','Cabin','Embarked'], axis = 1)
dfTestX.loc[np.isnan(dfTestX['Age']), 'Age'] = np.mean(dfTestX['Age'])
dfTestX.loc[np.isnan(dfTestX['Fare']), 'Fare'] = np.mean(dfTestX['Fare'])

arTestX = np.array(dfTestX).T
dfTestY = dfTestYOri.drop('PassengerId', axis = 1)
arTestY = np.array(dfTestY).T.reshape([1, -1])
```

```python
dpTrainX = dataPool(data = arTrainX, axis = 1)
dpTrainY = dataPool(data = arTrainY.reshape([1,-1]), axis = 1)
dpTestX = dataPool(data = arTestX, axis = 1)
dpTestY = dataPool(data = arTestY.reshape([1,-1]), axis = 1)
```

```python
Xrange = [np.min(arTrainX, axis = 1).reshape([-1, 1]), np.max(arTrainX, axis = 1).reshape([-1, 1])]
arTrainX =  (arTrainX - Xrange[0])/ (Xrange[1] - Xrange[0]) * 2 - 1
arTestX = (arTestX - Xrange[0]) / (Xrange[1] - Xrange[0]) * 2 - 1

pca = PCA(n_components = 5)
pca.fit(X = arTrainX.T, y = arTrainY.T)
arTrainX = pca.transform(arTrainX.T).T
print(pca.explained_variance_ratio_)
arTestX = pca.transform(arTestX.T).T

```

```python
layersize = [  20, 1]
layercount = len(layersize)

seed = time.time()
X = tf.placeholder(shape = [arTrainX.shape[0], None], dtype = 'float')
Y_ = tf.placeholder(shape = [layersize[-1], None], dtype = 'float')
W = dict()
b = dict()
Y = dict()
Y[0] = X
layersizeTemp = [arTrainX.shape[0]] + layersize
for i in range(layercount):
    W[i+1] = tf.Variable(tf.random_normal(seed = np.random.randint(seed), shape = [layersizeTemp[i+1], layersizeTemp[i]]) * 0.01)
    b[i+1] = tf.Variable(tf.zeros(shape = [layersizeTemp[i+1], 1]))
    if i != layercount - 1:
        Y[i+1] = tf.nn.sigmoid(tf.matmul(W[i+1], Y[i]) + b[i+1])
Y[layercount] = tf.nn.sigmoid(tf.matmul(W[layercount], Y[layercount-1]) + b[layercount])
Youtput = Y[layercount]

correct_prediction = tf.equal(tf.round(Youtput), Y_)
correct_rate = tf.reduce_mean(tf.cast(correct_prediction, 'float'))

cost = tf.reduce_mean(tf.square(Youtput - Y_))
#cost = - tf.reduce_mean(Y_*tf.log(Youtput) + (1 - Y_)*tf.log(1 - Youtput))
```

```python
sess = tf.Session(config=tf.ConfigProto(log_device_placement=True))
init = tf.global_variables_initializer()
sess.run(init) 
```

```python
sess.run(W[1], feed_dict = {X:arTrainX, Y_:arTrainY})
```

```python
#train_step = tf.train.AdagradOptimizer(1).minimize(crossEntropy)
train_step = tf.train.GradientDescentOptimizer(0.5).minimize(cost)
for i in range(50001):
    #sess.run(train_step, feed_dict = {X:dpTrainX.getData(size=50), Y_:dpTrainY.getData(size=50)})
    sess.run(train_step, feed_dict = {X:arTrainX, Y_:arTrainY})
    if i % 1000 == 0:
        print(calculatePrecision(arTrainX, arTrainY, correct_rate), calculatePrecision(arTestX, arTestY, correct_rate))
```
