```python
dfTrainOri = pd.read_csv(filepath_or_buffer = 'data/train.csv')
dfTestOri = pd.read_csv(filepath_or_buffer = 'data/test.csv')
dfTestYOri = pd.read_csv(filepath_or_buffer = 'data/gender_submission.csv')
```

```python
dfWhole = pd.concat([dfTrainOri, dfTestOri])
dfWhole.index = dfWhole['PassengerId']
dfWhole = dfWhole.drop('PassengerId', axis = 1)
dfWhole
```

```python
dfNames = pd.DataFrame(index = dfWhole.index)
names = list(map(lambda x:x.split('('), dfWhole['Name']))
names = list(map(lambda x:len(x)==2 and [x[0],x[1].replace(')','')] or x, names))
dfNames['MaidenName'] = list(map(lambda x:len(x)==2 and x[1] or '', names))
names = list(map(lambda x:x[0].split(','), names))
dfNames['FamilyName'] = list(map(lambda x:x[0], names))
names = list(map(lambda x:x[1], names))
names = list(map(lambda x:x[1:].split('.'), names))
dfNames['type'] = list(map(lambda x:x[0], names))
names = list(map(lambda x:x[1][1:], names))
names = list(map(lambda x:x.split(' '), names))
dfNames['FirstName'] = list(map(lambda x:x[0], names))
dfNames['MiddleName'] = list(map(lambda x:len(x)>1 and x[1] or '', names))
```

```python
# /******************************* 分析各种分组的组内样本死亡方差 *********************************/
# /************************** 同一张 ticket 的乘客同生同死可能性很高 ******************************/
dfTrain = dfWhole.loc[~pd.isnull(dfWhole['Survived']), :]
dfTrainNames = dfNames.loc[~pd.isnull(dfWhole['Survived']), :]
#计算整个Train集的总死亡率方差
deathRateVarSum = 0
dfGroup = dfTrain.loc[:, 'Survived']
deathRateVar = np.var(dfGroup)
deathRateVarSum += deathRateVar * dfGroup.shape[0]
print("Death rate variance of sum of the whole train dataset: %f"%deathRateVarSum)

#针对每个TicketGroup计算死亡率方差，然后整合
grpsByTicket = dfTrain.groupby('Ticket').groups
calculateSumOfGroupinDeathVars(grpsByTicket, dfTrain)

#针对每个TicketClassGroup计算死亡率方差，然后整合
grpsByPclass = dfTrain.groupby('Pclass').groups
calculateSumOfGroupinDeathVars(grpsByPclass, dfTrain)

#针对每个FamilyNameGroup计算死亡率方差，然后整合
grpsByFamilyName = dfTrainNames.groupby('FamilyName').groups
calculateSumOfGroupinDeathVars(grpsByFamilyName, dfTrain)

```

```python
# /************************** 增加特征：每位乘客同ticket内其他乘客已知生存率 ********************************/

grpsByTicketWhole = dfWhole.groupby('Ticket').groups
for ticket in grpsByTicketWhole:
    surviveRateAverage = np.mean(dfWhole.loc[grpsByTicketWhole[ticket], 'Survived'])
    dfWhole.loc[grpsByTicketWhole[ticket], 'IngroupSurviveRateOfTicket'] = surviveRateAverage
dfWhole.loc[pd.isnull(dfWhole['IngroupSurviveRateOfTicket']), 'IngroupSurviveRateOfTicket'] = np.mean(dfWhole['IngroupSurviveRateOfTicket'])
```

```python
def calculateSumOfGroupinDeathVars(groups, dfTrain):
    deathRateVarSum = 0
    for member in groups:
        deathRateVar = np.var(dfTrain.loc[groups[member], 'Survived'])
        deathRateVarSum += deathRateVar * len(groups[member])
    print("Death rate variance of sum of groups(by pclass) of train dataset: %f"%deathRateVarSum)
```

```python
def checkIfTicketSameInafamily(dfWhole, dfNames):
    dctTickets = dfWhole.groupby('Ticket').groups
    count = 0
    for ticket in dctTickets:
        dfTicketFamilyNames = dfNames.loc[dctTickets[ticket], 'FamilyName']
        if np.sum(dfTicketFamilyNames != dfTicketFamilyNames.iloc[0]) > 0:
            print("Found people with different FamilyNames in ticket %s"%(ticket))
            print(dfWhole.loc[dctTickets[ticket], ['Name', 'Ticket', 'Parch', 'SibSp']])
            count += 1
    print(count)
```
